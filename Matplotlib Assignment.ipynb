```python
Maxi = [39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25]
Min = [21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18]

months=['Jan','Fab','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

plt.plot(months, Maxi, 'ro')
plt.plot(months, Min, 'go')
plt.xlabel('Months')
plt.ylabel('MIn & Max Temp')
```

```python
titanic= pd.read_csv('https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv')
titanic.head()
```

```python
df=titanic.groupby('sex').count()
df
```

```python
df.pclass
```

```python
df.plot.pie(y='pclass', autopct='%.2f%%', shadow=True, explode=[.05,0], fontsize=30, figsize=(8,8), labeldistance=.9)
```

```python
# Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

sns.scatterplot(x=titanic.age, y=titanic.fare, hue=titanic.sex , s=80)
```
