### spliting the data into test and train

```python
# split the data into independent (X) and dependent (Y) data sets

X = bost.iloc[: , 0:12].values           # contains the features
Y = bost.iloc[: , 13].values              # contains the result
```

```python
# spliting the data into test and train dataset(75%-train and 25%-test)

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state= 0)
```

```python
# scale the data (Feature scaling)

from sklearn.preprocessing import StandardScaler

sc= StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
# Linear regression

from sklearn.linear_model import LinearRegression

lin = LinearRegression()
lin.fit(X_train, Y_train)
```

```python
# prediction

pred=lin.predict(X_test)
```

```python
# mean square error(MSE)

from sklearn.metrics import mean_squared_error

mean_squared_error(Y_test, pred )
```

```python
# training accuracy

print('Linear Regression Training Accuracy:' , lin.score(X_train, Y_train))
```

```python
# R2

from sklearn.metrics import r2_score

score= r2_score(Y_test, pred)
score
```
